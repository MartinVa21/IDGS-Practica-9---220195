#### Sesión 6: Análisis Específico de TV Shows

```python
# Análisis de países que producen TV Shows
countries = {}
netflix_shows['country'] = netflix_shows['country'].fillna('Unknown')
cou = list(netflix_shows['country'])

# Procesar países (algunos títulos tienen múltiples países)
for i in cou:
    i = list(i.split(','))
    if len(i) == 1:
        if i in list(countries.keys()):
            countries[i] += 1
        else:
            countries[i[0]] = 1
    else:
        for j in i:
            if j in list(countries.keys()):
                countries[j] += 1
            else:
                countries[j] = 1

# Limpiar datos de países
countries_fin = {}
for country, no in countries.items():
    country = country.replace(' ', '')
    if country in list(countries_fin.keys()):
        countries_fin[country] += no
    else:
        countries_fin[country] = no

# Ordenar países por cantidad de contenido
countries_fin = {k: v for k, v in sorted(countries_fin.items(), key=lambda item: item[1], reverse=True)}

# Visualizar top 10 países productores de TV Shows
plt.figure(figsize=(8, 8))
ax = sns.barplot(x=list(countries_fin.keys())[0:10], y=list(countries_fin.values())[0:10])
ax.set_xticklabels(list(countries_fin.keys())[0:10], rotation=90)
plt.title("Top 10 Países Productores de TV Shows")
plt.show()
```

### Sistema de recomendacion basado en contenido

Recomendador basado en la descripción de la trama (Recomendaciones basadas en contenido)

```python
#Descripción de la trama basada en recomendador (Recomendaciones basadas en contenido)

netflix_dataset['description'].head()
```

```python
#Sistema de Recomendación (Basado en Contenido)
#Importar TfIdfVectorizer de scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

#Define un objeto Vectorizador TF-IDF. Elimina todas las palabras vacías en inglés como 'the', 'a', etc.
tfidf = TfidfVectorizer(stop_words='english')

#Reemplazar NaN con un string vacío
netflix_dataset['description'] = netflix_dataset['description'].fillna('')

#Construya la matriz TF-IDF requerida ajustando y transformando los datos.
tfidf_matrix = tfidf.fit_transform(netflix_dataset['description'])

#Muestra la forma de tfidf_matrix
tfidf_matrix.shape
```

```python
#Importar kernel_lineal
from sklearn.metrics.pairwise import linear_kernel

# Calcular la matriz de similitud coseno
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Construir un mapa inverso de índices y títulos de películas
indices = pd.Series(netflix_dataset.index, index=netflix_dataset['title']).drop_duplicates()
```

```python
# Función que toma como entrada el título de una película y devuelve las películas más similares
def get_recommendations(title, cosine_sim=cosine_sim):
    # Obtener el índice de la película que coincide con el título dado
    idx = indices[title]

    # Obtener las puntuaciones de similitud (coseno) entre esa película y todas las demás
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Ordenar las películas según su puntuación de similitud (de mayor a menor)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Obtener las puntuaciones de las 10 películas más similares (excluyendo la misma película)
    sim_scores = sim_scores[1:11]

    # Obtener los índices de esas 10 películas
    movie_indices = [i[0] for i in sim_scores]

    # Devolver los títulos de las 10 películas más similares
    return netflix_dataset['title'].iloc[movie_indices]
```

```python
#Llenando los valores nulos con una cadena vacía.
filledna=netflix_dataset.fillna('')
filledna.head()
```

```python
#Limpiando los datos - poniendo todas las palabras en minúsculas
def clean_data(x):
        return str.lower(x.replace(" ", ""))
```

```python
#Identificación de características en las que se debe filtrar el modelo.
features=['title','director','cast','listed_in','description']
filledna=filledna[features]
```

```python
for feature in features:
    filledna[feature] = filledna[feature].apply(clean_data)
    
filledna.head()
```

```python
def create_soup(x):
    return x['title']+ ' ' + x['director'] + ' ' + x['cast'] + ' ' +x['listed_in']+' '+ x['description']

filledna['soup'] = filledna.apply(create_soup, axis=1)
```

```python
# Importa CountVectorizer y crea la matriz de conteo
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(filledna['soup'])
```

```python
# Calcula la matriz de similitud coseno basada en la matriz de conteo.
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
# Restablecer el índice de nuestro DataFrame principal y construir la asignación inversa como antes.
filledna=filledna.reset_index()
indices = pd.Series(filledna.index, index=filledna['title'])
```

```python
# Función que toma como entrada el título de una película y devuelve las 10 películas más similares
def get_recommendations_new(title, cosine_sim=cosine_sim):
    # Normaliza el título: elimina espacios y convierte a minúsculas
    title = title.replace(' ', '').lower()

    # Obtiene el índice de la película que coincide con el título normalizado
    idx = indices[title]

    # Obtiene las puntuaciones de similitud coseno entre esa película y todas las demás
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Ordena las películas según su puntuación de similitud (de mayor a menor)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Toma las 10 películas más similares (excluye la misma película original)
    sim_scores = sim_scores[1:11]

    # Obtiene los índices de esas 10 películas
    movie_indices = [i[0] for i in sim_scores]

    # Devuelve los títulos de las 10 películas más similares
    return netflix_dataset['title'].iloc[movie_indices]
```

### Sistema de recomendación basado en contenido para el segundo DataSet

```python
netflix_dataset2['description'].head()
```

```python
# Importar el vectorizador TF-IDF desde la biblioteca scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

# Definir un objeto TF-IDF Vectorizer.
# Este eliminará todas las palabras vacías en inglés (como 'the', 'a', etc.)
tfidf = TfidfVectorizer(stop_words='english')

# Reemplazar los valores nulos (NaN) en la columna 'description' con una cadena vacía
netflix_dataset2['description'] = netflix_dataset2['description'].fillna('')

# Construir la matriz TF-IDF ajustando y transformando la columna 'description'
tfidf_matrix = tfidf.fit_transform(netflix_dataset2['description'])

# Mostrar la forma (dimensiones) de la matriz TF-IDF resultante
tfidf_matrix.shape
```

```python
#Importar kernel_lineal
from sklearn.metrics.pairwise import linear_kernel

# Calcular la matriz de similitud coseno
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Construir un mapa inverso de índices y títulos de películas
indices = pd.Series(netflix_dataset2.index, index=netflix_dataset2['title']).drop_duplicates()
```

Función que toma el título de la película como entrada y genera la mayoría de las películas similares

```python
# Función que recibe el título de una película o serie y devuelve las 10 más similares
def get_recommendations(title, cosine_sim=cosine_sim):
    # Obtiene el índice de la película o serie que coincide con el título proporcionado
    idx = indices[title]

    # Obtiene las puntuaciones de similitud coseno entre esa película/serie y todas las demás
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Ordena las películas/series por su puntuación de similitud (de mayor a menor)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Toma las puntuaciones de las 10 más similares (omitiendo la primera, que es la misma)
    sim_scores = sim_scores[1:11]

    # Extrae los índices de esas 10 películas o series
    movie_indices = [i[0] for i in sim_scores]

    # Devuelve los títulos de las 10 películas o series más similares
    return netflix_dataset2['title'].iloc[movie_indices]
```

```python
get_recommendations('Inception')
```

```python
get_recommendations('Avengers: Infinity War')
```

```python
get_recommendations('Dil Dhadakne Do')
```

### Filtrado basado en contenido según múltiples métricas

```python
#Rellenar valores nulos con una cadena vacía.
filledna=netflix_dataset2.fillna('')
filledna.head()
```

```python
#Limpieza de datos: convertir todas las palabras en minúsculas
def clean_data(x):
        return str.lower(x.replace(" ", ""))
```

```python
#Identificar las características sobre las que se filtrará el modelo.
features=['title','director','cast','description']
filledna=filledna[features]
```

```python
for feature in features:
    filledna[feature] = filledna[feature].apply(clean_data)
    
filledna.head()
```

```python
def create_soup(x):
    return x['title']+ ' ' + x['director'] + ' ' + x['cast'] +' '+ x['description']

filledna['soup'] = filledna.apply(create_soup, axis=1)
```

```python
# Importa CountVectorizer y crea la matriz de conteo
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(filledna['soup'])
```

```python
# Calcule la matriz de similitud de coseno basándose en count_matrix
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
# Restablecer el índice de nuestro DataFrame principal y construir un mapeo inverso como antes
filledna=filledna.reset_index()
indices = pd.Series(filledna.index, index=filledna['title'])
```

```python
# Función que toma como entrada el título de una película o serie
# y devuelve las 10 más similares según la similitud del coseno
def get_recommendations_new(title, cosine_sim=cosine_sim):
    # Normaliza el título: elimina espacios y lo convierte a minúsculas
    title = title.replace(' ', '').lower()

    # Obtiene el índice de la película o serie que coincide con el título normalizado
    idx = indices[title]

    # Calcula las puntuaciones de similitud coseno entre esta película/serie y todas las demás
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Ordena las películas/series por su similitud (de mayor a menor)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Toma las 10 más similares, ignorando la primera (que es la misma película/serie)
    sim_scores = sim_scores[1:11]

    # Extrae los índices de las películas/series similares
    movie_indices = [i[0] for i in sim_scores]

    # Devuelve los títulos de las 10 películas o series más similares
    return netflix_dataset2['title'].iloc[movie_indices]
```
